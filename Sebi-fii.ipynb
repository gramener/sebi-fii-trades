```python
#data['TR_DATE'].head()
x = data.groupby('ISIN').apply(lambda v: v.set_index('TR_DATE').resample('M',how='sum'))
x = x.reset_index()
piv = x.pivot_table(rows='ISIN',cols='TR_DATE', values='VALUE (in Rs)')
```

Find scrip names with maximum activity, less-activity-but-spikes.

```python
vc = piv.apply(lambda x: len(x.dropna()),axis=1)
vc = vc.reset_index()
vc.rename(columns={0:'COUNTS'},inplace=True)
vc = vc.set_index('ISIN')
piv = piv.merge(vc, left_index=True, right_index=True)
piv = piv.sort('COUNTS',ascending=False)
```

Let's look at the types of transactions that have happened during this duration. The <b>quantity of equity traded</b> and <b>value of these</b> should shed some light on, the behavior of the transactions.

```python
tr = data['TR_TYPE(*)'].value_counts()
tr = tr.reset_index().rename(columns={'index':'trtype',0:'counts'})
tr['qty'] = tr['trtype'].apply(lambda v: data[data['TR_TYPE(*)']==v]['QUANTITY'].sum())
tr['value'] = tr['trtype'].apply(lambda v: data[data['TR_TYPE(*)']==v]['VALUE (in Rs)'].sum())
tr['avg value/unit'] = tr['value']/tr['qty']
tr.replace({'trtype':trans.to_dict()['Nature of transaction']})
```
